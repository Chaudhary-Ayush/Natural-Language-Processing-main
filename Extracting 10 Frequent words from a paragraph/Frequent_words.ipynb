```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from wordcloud import WordCloud
from matplotlib import pyplot as plt
```

```python
with open("E:\\6th semester\\College\\NLP lab\\next exp\\ADHD.txt", 'r') as file:
    text = file.read()
```

```python
# Tokenize the text into words
words = word_tokenize(text)
```

```python
# Removing punctuation and lowercase all words
words = [word.lower() for word in words if word.isalnum()]
```

```python
# Calculating the frequency distribution of words
fdist = FreqDist(words)
```

```python
# Printing the most common words
print("Top 10 frequent words:")
for word, frequency in fdist.most_common(10):  
    print(f"{word}: {frequency}")
```

```python
#creating word cloud
word_cloud1 = WordCloud(collocations = False, background_color = 'Black',
                        width = 1050, height = 1080).generate(text)
```

```python
plt.imshow(word_cloud1, interpolation='quadric')
plt.axis("off")
plt.show()
```
